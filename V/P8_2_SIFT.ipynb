```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img1 = cv2.imread('/content/2.jpg')

# Convert the image to grayscale
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

# Create SIFT detector object
sift = cv2.SIFT_create()

# Detect keypoints and compute descriptors
keypoints_1, descriptors_1 = sift.detectAndCompute(gray1, None)

# Draw keypoints on the image
img_1 = cv2.drawKeypoints(img1, keypoints_1, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# Plotting the image with SIFT keypoints
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB))
plt.title('Image with SIFT Keypoints')
plt.axis('on')  # Show axis
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load images
img1 = cv2.imread('/content/1.jpg')
img2 = cv2.imread('/content/2.jpg')

# Convert images to grayscale
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# Initialize SIFT detector
sift = cv2.SIFT_create()

# Detect keypoints and compute descriptors for both images
keypoints_1, descriptors_1 = sift.detectAndCompute(gray1, None)
keypoints_2, descriptors_2 = sift.detectAndCompute(gray2, None)

# Initialize Brute-Force Matcher
bf = cv2.BFMatcher()

# Match descriptors
matches = bf.match(descriptors_1, descriptors_2)

# Sort matches by distance (lower is better)
matches = sorted(matches, key=lambda x: x.distance)

# Draw matches
img_matches = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:10], None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

# Convert image to RGB format for displaying with matplotlib
img_matches_rgb = cv2.cvtColor(img_matches, cv2.COLOR_BGR2RGB)

# Display the images with matches
plt.figure(figsize=(16, 8))
plt.imshow(img_matches_rgb)
plt.title('SIFT Matches Between Image 1 and Image 2')
plt.axis('on')  # Show axis
plt.show()

```
