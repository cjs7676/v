```python
import numpy as np
from google.colab.patches import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/cricket.png')
print(img)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
haar_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
faces_rect = haar_cascade.detectMultiScale(gray_img, 1.1, 9)
for(x, y, w, h) in faces_rect:
  cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
#haar_cascade = cv2.CascadeClassifier('/content/haarcascade_eye.xml')
#eyes_rect = haar_cascade.detectMultiScale(gray_img, 1.1, 9)
#for(x, y, w, h) in eyes_rect:
  #cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
cv2_imshow(img)
cv2.waitKey(0)
```

```python
from google.colab.patches import cv2
import matplotlib.pyplot as plt
img = cv2.imread('/content/images.jfif')
edges = cv2.Canny(img, 100, 200, 3, L2gradient = True)
plt.figure()
plt.title('fox')
plt.imsave('test.png', edges, cmap = 'gray', format = 'png')
plt.imshow(edges, cmap = 'gray')
plt.show()
```

```python
from google.colab.patches import cv2
import numpy as np
image = cv2.imread('/content/images.jfif')
cv2.waitKey(0)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
edged = cv2.Canny(gray, 30, 200)
cv2.waitKey(0)
contours, hierarchy = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
cv2_imshow(edged)
cv2.waitKey(0)
print("Number of Contours found = " + str(len(contours)))
cv2.drawContours(image, contours, -1, (0, 255, 0), 3)
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
