```python
import cv2
import matplotlib.pyplot as plt

# Path to the image
path = r'/content/krishna.jpg'

# Loading the image
image = cv2.imread(path)

# Resize operations
size1 = cv2.resize(image, (0, 0), fx=0.1, fy=0.1)
size2 = cv2.resize(image, (1050, 1610))
stretch_near = cv2.resize(image, (780, 540), interpolation=cv2.INTER_LINEAR)

# Titles and images
Titles = ["Original", "Size1", "Size2", "Size3"]
images = [image, size1, size2, stretch_near]

# Display the original image
plt.subplot(2, 2, 1)
plt.title(Titles[0])
plt.imshow(cv2.cvtColor(images[0], cv2.COLOR_BGR2RGB))
plt.axis('on')

# Display size1
plt.subplot(2, 2, 2)
plt.title(Titles[1])
plt.imshow(cv2.cvtColor(images[1], cv2.COLOR_BGR2RGB))
plt.axis('on')

# Display size2
plt.subplot(2, 2, 3)
plt.title(Titles[2])
plt.imshow(cv2.cvtColor(images[2], cv2.COLOR_BGR2RGB))
plt.axis('on')

# Display stretch_near
plt.subplot(2, 2, 4)
plt.title(Titles[3])
plt.imshow(cv2.cvtColor(images[3], cv2.COLOR_BGR2RGB))
plt.axis('on')

plt.show()
```
