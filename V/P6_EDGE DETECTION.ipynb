```python
import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('/content/shiv.jpg')

# Convert the image from BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert the RGB image to grayscale
gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

# Apply thresholding to get a binary image
ret, thresh = cv2.threshold(gray, 125, 255, 0)

# Find contours in the binary image
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Create a copy of the RGB image to draw contours on
copy_img = img_rgb.copy()
cv2.drawContours(copy_img, contours, -1, (0, 0, 255), 2)

# Create a figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 10))

# Display the original RGB image
axes[0].imshow(img_rgb)
axes[0].set_title('Original Image')
axes[0].axis('on')  # Show axis

# Display the grayscale image
axes[1].imshow(gray, cmap='gray')
axes[1].set_title('Grayscale Image')
axes[1].axis('on')  # Show axis

# Display the image with contours
axes[2].imshow(copy_img)
axes[2].set_title('Image with Contours')
axes[2].axis('on')  # Show axis

# Show the figure
plt.show()
```
